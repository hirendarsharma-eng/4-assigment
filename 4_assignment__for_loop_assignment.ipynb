## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
n = int(input("Enter number: "))
flag =0
if n<=1:
  flag =1
else:
  for i in range(2,n-1):
    if n%i==0:
      flag=1
      break
    else:
      flag = 0
if flag==0:
  print("prime")
else:
  print("No prime")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
n = int(input("Enter n value: "))

for num in range(2, n + 1):
    flag = 1
    for i in range(2, num):
        if num % i == 0:
            flag = 0
            break
    if flag == 1:
        print(num)


```
